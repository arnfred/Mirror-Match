```python
import display
import weightMatrix
import matching
import louvainmatch
import louvain
import features
import numpy
import scoring
import cv2
from itertools import combinations
```

```python
paths = ['../../images/gallagher/partial_match/100_1941_small.jpg', '../../images/gallagher/partial_match/100_1942_small.jpg']
```

```python
# Get images
images = map(features.loadImage, paths)
keypoint_type = "SIFT"
descriptor_type = "SIFT"

# Get keypoints
indices, ks, ds = features.getFeatures(paths, keypoint_type, descriptor_type)
pos = features.getPositions(ks)
positions = numpy.array(pos)
```

```python
def getHomographyPath(paths) :
    d = "/".join(paths[0].split('/')[0:-1])
    p1 = paths[0].split('/')[-1].split(".")[0]
    p2 = paths[1].split('/')[-1].split(".")[0]
    return "%s/%s_%s_hom.npy" % (d, p1, p2)
homography_path = getHomographyPath(paths)
with open(homography_path) :
    homography = numpy.loadtxt(homography_path)
```

```python
ds0_len = len(ds[indices == 0])
ds1_len = len(ds[indices == 1])
distances = numpy.zeros((ds0_len, ds1_len))
for i, d0 in enumerate(ds[indices == 0]) :
    for j, d1 in enumerate(ds[indices == 1]) :
        distances[i,j] = numpy.linalg.norm(d1-d0)
distances
```

```python
best_dist = numpy.argsort(numpy.reshape(distances, ds0_len*ds1_len))
print(best_dist)
def get_pair(idx) :
    idx_row = idx / ds1_len
    idx_col = idx % ds1_len
    return (idx_row, idx_col)
```

```python
i,j = get_pair(best_dist[0])
print(distances[i,j])
print(distances.min())
```

```python
nb_matches = 100
best_n_matches = map(get_pair, best_dist[:200])
```

```python
(pos_im1, pos_im2) = (features.getPositions(ks[indices == 0]), features.getPositions(ks[indices == 1]))
match_pos, ss, uu = zip(*[((pos_im1[i],pos_im2[j]), distances[i,j], 0) for (i,j) in best_n_matches])
true_distances = [matching.matchDistance(p1,p2, homography) for p1,p2 in match_pos]
display.matchPoints(images[0], images[1], match_pos, true_distances, filename="naive-match.png", line_width=2)
```
