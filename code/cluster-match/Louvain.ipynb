```python
indices, keypoints, descriptors = featureGraph.getDescriptors(paths)
print("Number of keypoints: %i" % (len(keypoints)))
images = map(features.loadImage, paths)
labels = map(features.getLabel, paths)
print(labels)

# Get weights
full_weights = weightMatrix.init(descriptors)
score_weights = weightMatrix.pruneTreshold(full_weights, 20)
cluster_weights = weightMatrix.pruneHighest(score_weights, 3)

# Get graphs
cluster_graph = featureGraph.init(cluster_weights)
score_graph = featureGraph.init(score_weights)
```

```python
import louvain

partition = louvain.cluster(cluster_weights, True)
p = featureGraph.setPartition(cluster_graph, partition)
```

```python
featureGraph.setFeatureIndices(cluster_graph, indices)
featureGraph.setFeatureIndices(score_graph, indices)
display.draw_graph(cluster_graph, p)
```

```python
# Show on image
c = [11]
featureGraph.setPositions(score_graph, keypoints)

one_cluster = featureGraph.getPartition(score_graph, p, c)
display.graph_on_images(one_cluster, images, p)
```

```python
scores, partition_indices = scoring.scoreWeights(score_weights, partition, numpy.array(indices), verbose=True)
if (len(scores) > 3) : print("score: %0.4f" % (sum(scores)))
else : print("Can't match images based on 4 or less traits")
```

```python
display.trait_graph(score_graph, p, partition_indices, scores, images)
```
