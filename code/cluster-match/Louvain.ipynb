```python
import os
import fnmatch
import features
import cluster
import numpy
from random import sample
import graph_tool.all as gt


# testset = "../../images/testsets/WC818C/"
testset = "../../images/testsets/small_test/"

# Get a list of all images
all_paths = [testset + f for (_,_,filenames) in os.walk(testset) for f in fnmatch.filter(filenames, "*.png")]
l1 = 0
l2 = 1
while (l1 != l2) :
    paths = sample(all_paths, 2)
    l1 = features.getLabel(paths[0])
    l2 = features.getLabel(paths[1])
print(paths)
#paths = ['../../images/testsets/small_test/Bill_Gates_0003.png', '../../images/testsets/small_test/Bill_Gates_0007.png']
#paths = ['../../images/testsets/small_test/Anna_Kournikova_0012.png', '../../images/testsets/small_test/Anna_Kournikova_0008.png']
#paths = ['../../images/testsets/small_test/Ann_Veneman_0003.png', '../../images/testsets/small_test/Bill_Gates_0002.png']
#paths = ['../../images/testsets/small_test/Anna_Kournikova_0006.png', '../../images/testsets/small_test/Bill_Gates_0002.png']
#paths = ['../../images/testsets/small_test/Ann_Veneman_0002.png', '../../images/testsets/small_test/Ann_Veneman_0004.png']
#paths = ['../../images/testsets/small_test/Ann_Veneman_0003.png', '../../images/testsets/small_test/Ann_Veneman_0008.png'] # Not matched, because only two traits. The traits were good though
#paths = ['../../images/testsets/small_test/Bill_Gates_0008.png', '../../images/testsets/small_test/Bill_Gates_0006.png'] # Not matched, but when I lower treshold it get's a big negative score
#paths = ['../../images/testsets/small_test/Ann_Veneman_0009.png', '../../images/testsets/small_test/Anna_Kournikova_0012.png'] # A match that makes sense but shouldn't happen
#paths = ['../../images/testsets/small_test/Bill_Gates_0002.png', '../../images/testsets/small_test/Bill_Gates_0005.png'] # correctly matched, however they shuldn't have been
#paths = ['../../images/testsets/small_test/Ann_Veneman_0001.png', '../../images/testsets/small_test/Ann_Veneman_0002.png'] # matches better when treshold is on 1.4
indices, keypoints, descriptors = cluster.getDescriptors(paths)
images = map(features.loadImage, paths)
labels = map(features.getLabel, paths)
print(labels)

connected, weights = cluster.initGraph(descriptors, indices)
connected_light = cluster.prune(connected, weights, 1.4)
filtered = cluster.prune(connected, weights, 2.3)
```

```python
import louvain
p = louvain.cluster(filtered)
```

```python
from itertools import groupby
t = [(k, len(list(g))) for k,g in groupby(sorted(p.fa))]
#print(t)
def reassign(partitioning) :
    assignments = [(i,k) for i,(k,g) in enumerate(groupby(sorted(partitioning.fa)))]
    p_new = p.copy()
    for i,k in assignments :
        p_new.fa[p.fa == k] = i
    return p_new

p = reassign(p)
#print(t)
#print(p.fa)
```

```python
cluster.show(filtered, p)
```

```python
# Show on image
c = [61]
cluster.setPositions(connected_light, keypoints)
connected_cluster = cluster.getPartition(connected_light, p, c)
cluster.showOnImages(connected_cluster, images, p)
```
