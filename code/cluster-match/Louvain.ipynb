```python
indices, keypoints, descriptors = cluster.getDescriptors(paths)
print("Number of keypoints: %i" % (len(keypoints)))
images = map(features.loadImage, paths)
labels = map(features.getLabel, paths)
print(labels)

connected, weights = cluster.initGraph(descriptors, indices, 1.1)
```

```python
connected_light = cluster.prune(connected, weights, edges_per_vertex=20)

filtered = cluster.prune(connected, weights, edges_per_vertex=4.5)
#bipartite = cluster.removeIntraImageEdges(connected, weights)
#filtered = cluster.prune(bipartite, weights, edges_per_vertex=4.5)

#fraction_b = float(bipartite.num_edges()) / float(bipartite.num_vertices()) 
#fraction_o = float(connected.num_edges()) / float(connected.num_vertices()) 
fraction_f = float(filtered.num_edges()) / float(filtered.num_vertices()) 
fraction_c = float(connected_light.num_edges()) / float(connected_light.num_vertices()) 
#print("Number of edges per vertex in connected: %f.2" % fraction_o + " (%i edges)" % connected.num_edges())
#print("Number of edges per vertex in bipartite: %f.2" % fraction_b + " (%i edges)" % bipartite.num_edges())
print("Number of edges per vertex in filtered: %f.2" % fraction_f + " (%i edges)" % filtered.num_edges())
print("Number of edges per vertex in connected_light: %f.2" % fraction_c + " (%i edges)" % connected_light.num_edges())
```

```python
import louvain

p, nb_clusters = louvain.cluster(filtered)
```

```python
p = cluster.reassign(p)
```

```python
display.draw_graph(filtered, p)
```

```python
# Show on image
c = [11]
cluster.setPositions(connected_light, keypoints)
# Create bipartite graph
connected_cluster = cluster.getPartition(connected_light, p, c)
display.graph_on_images(connected_cluster, images, p)
```

```python
scores, partition_indices = cluster.scoreAll(connected_light, p, nb_clusters, verbose=True)
if (len(scores) > 2) : print("score: %0.4f" % (sum(scores)))
else : print("Can't match images based on 2 or less traits")
```

```python
display.trait_graph(connected_light, p, partition_indices, scores, images)
```
