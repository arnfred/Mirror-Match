```python
import features
import display
import os
import mirrormatch
import pylab
import cv2
```

```python
# Rename the photos to numbers
def renamePics(folder, img_type = "jpg") :
    new_paths = []
    images = [im for im in sorted(os.listdir(folder)) if im[-4:].lower() == "." + img_type]
    for i, img in enumerate(images) :
        old_path = folder + img
        new_path = folder + str(i+1) + "." + img_type
        os.rename(old_path, folder + str(i+1) + "." + img_type)
        new_paths.append(new_path)
    return new_paths
```

```python
# Given two photos, find potential matches
def findMatches(img_paths, indices = (0, 1)) :
    paths = (img_paths[indices[0]], img_paths[indices[1]])
    match_fun = mirrormatch.match(paths, { "verbose" : True, "ratio" : ("other", "all") })
    return paths, match_fun
```

```python
def getHomographyPath(paths) :
    d = "/".join(paths[0].split('/')[0:-1])
    p1 = paths[0].split('/')[-1].split(".")[0]
    p2 = paths[1].split('/')[-1].split(".")[0]
    return "%s/%s_%s_hom.npy" % (d, p1, p2)
```

```python
# Show n matches with indices
def show_matches(images, match_fun, page_size = 20, color = '#00aaff', img_separation = 20, line_width = 0.8, threshold = 0.8) :
    
    # get matches
    matches = match_fun(threshold)
    
    # get page matches
    page_idx = list(set(numpy.random.randint(0, len(matches[0]), page_size)))
    match_page = numpy.array(matches[0])[page_idx]

    # Construct unified image
    im3 = display.appendimages(images[0], images[1], img_separation)

    # Create figure
    fig = pylab.figure(frameon=False, figsize=(12.0, 8.0))
    ax = pylab.Axes(fig, [0., 0., 1., 1.])

    ax.set_axis_off()
    fig.add_axes(ax)

    # Display image
    pylab.gray()
    ax.imshow(im3)
    
    # Plot all lines
    offset_x = images[0].shape[1]
    for i,((x1,y1),(x2,y2)) in zip(page_idx, match_page) :
        # Add line
        ax.plot([x1, x2+offset_x + img_separation], [y1,y2], color=color, lw=line_width)
        # Add number
        ax.text(x1, y1, str(i))
        ax.text(x2+offset_x + img_separation, y2, str(i))
    
    pylab.xlim(0,im3.shape[1])
    pylab.ylim(im3.shape[0],0)
    
    return matches
```

```python
def getHomography(paths, matches, images, selected = []) :
    correct = [m for i,m in enumerate(matches[0]) if i in selected]
    display.matchPoints(images[0], images[1], correct, filename=None)
    match_1, match_2 = zip(*correct)
    homography, blup = cv2.findHomography(numpy.array(match_1), numpy.array(match_2), cv2.RANSAC)
    homography_path = getHomographyPath(paths)
    numpy.savetxt(homography_path, homography)
    print(homography)
    print(homography_path)
```

```python
folder = "../../../city_set/center/"
indices = (0,1)
imagepaths = renamePics(folder)
print("Paths: %s, %s" % (imagepaths[indices[0]], imagepaths[indices[1]]))
paths, match_fun = findMatches(imagepaths, indices = indices)
images = map(features.loadImage, paths)
```

```python
matches = show_matches(images, match_fun, page_size = 10, threshold = 0.75)
```
