```python
# Get homography
def getHomographyPath(paths) :
    d = "/".join(paths[0].split('/')[0:-1])
    p1 = paths[0].split('/')[-1].split(".")[0]
    p2 = paths[1].split('/')[-1].split(".")[0]
    return "%s/%s_%s_hom.npy" % (d, p1, p2)
homography_path = getHomographyPath(paths)
with open(homography_path) :
    homography = numpy.loadtxt(homography_path)
```

```python
options = {
       "leaf_size": 10,
       "radius_size": 250,
       "dist_threshold": 300,
       "ratio_boost" : 1,
       "shuffle_keypoints" : False,
       "group_limit" : 4,
       "match_type" : "radius",
}
bm_fun  = ballmatch.match(paths, options)
```

```python
def accu(fp_fun, tp_fun) :
    for ut in numpy.linspace(0.0, 1.0, 100) :
        fp = fp_fun(ut)
        tp = tp_fun(ut)
        if len(fp) == 0 :
            percent_correct = 1.0
        else :
            percent_correct = len(tp) / float(len(fp) + len(tp))
        matches = len(tp) + len(fp)
        yield matches, percent_correct
```

```python
def penalize(g, g_limit, g_penalty) :
    if g > g_limit :
        return g_penalty
    else :
        return 1.0

def fp_tp(match_fun, homography, dt = 50, group_limit = 999, group_penalty = 1.0) :
    matches = numpy.array([(s, u, matching.matchDistance(p[0], p[1], homography), int(g)) for p, s, u, g in zip(*match_fun(1.0))])
    p = lambda g : penalize(g, group_limit, group_penalty)
    fp = [(s, u*p(g), g) for s,u,d,g in matches if d > dt]
    tp = [(s, u*p(g), g) for s,u,d,g in matches if d <= dt]
    fp_fun = lambda ut : numpy.array([(s,u,g) for (s,u,g) in fp if u < ut])
    tp_fun = lambda ut : numpy.array([(s,u,g) for (s,u,g) in tp if u < ut])
    return fp_fun, tp_fun
```

```python
def plot_accu(match_fun, group_limit = 999, group_penalty = 1.0, color = "#aa2200", label = None) :
    if label == None : label = "Penalty: %.2f, limit: %i" % (group_penalty, group_limit)
    fp, tp = fp_tp(match_fun, homography, dt = 50, group_limit = group_limit, group_penalty = group_penalty)
    matches_stat, correct_stat = zip(*list(accu(fp, tp)))
    plot(matches_stat, correct_stat, '-', color=color, label = label)
```
