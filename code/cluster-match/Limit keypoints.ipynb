```python
def getPartitionMatches(partitions, weights, indices, min_edges = 1, min_coherence = -1.0) :
    
    # index
    r = numpy.arange(0, weights.shape[0])
    # Get numpy array of indices
    for p in set(partitions) :
    
        partition_mask = partitions == p
        partition_weights = weights[partition_mask][:, partition_mask]
        for i,j in combinations(set(indices),2) :
    
            # Set up masks
            row_mask = partition_mask & (indices == i)
            col_mask = partition_mask & (indices == j)
            index_row = r[row_mask]
            index_col = r[col_mask]
    
            # Get weights
            pij_edges = weights[row_mask][:,col_mask]
            
            c = louvainmatch.getCoherence(partition_weights, partition_mask, indices, i, j)
            nb_e = numpy.sum(pij_edges > 0)
            if nb_e >= 1 and c > min_coherence:
                
                #pairs = [numpy.unravel_index(index, pij_edges.shape) for index in pij_edges.argsort(axis=None)[::-1]]
                sd0,sd1 = scoring.getPartitionDeviation(partition_mask, (indices == i, indices == j), ks)
                (m_i, m_j) = numpy.unravel_index(pij_edges.argmax(), pij_edges.shape)
                distance = matching.matchDistance(ks_s[index_row[m_i]].pt, ks_s[index_col[m_j]].pt, homography)
                print("%4i\tsd0: %.2f\tsd1: %.2f\tch: %.3f\tEdges: %i\tDistance: %.2f" % (p, sd0, sd1, c, nb_e, distance))
                yield (index_row[m_i], index_col[m_j])
```

```python
def cluster(weights, indices, split_limit = 10, prune_limit = 3, indent = "", level = 0, verbose = False) :
    partitions = louvain.cluster(weights, verbose=verbose)
    if level > 10 : return partitions
    p_set = set(partitions)
    
    r = numpy.arange(0, weights.shape[0])
    for p in p_set :
        partition_mask = partitions == p
        for i,j in combinations(set(indices),2) :
            # Set up masks
            row_mask = partition_mask & (indices == i)
            col_mask = partition_mask & (indices == j)
            index_row = r[row_mask]
            index_col = r[col_mask]
    
            # Get weights
            pij_edges = weights[row_mask][:,col_mask]
            p_edges = weights[partition_mask][:,partition_mask]
            nb_inter_edges = numpy.sum(pij_edges > 0)
            if numpy.sum(partition_mask) > split_limit :
                # Prune weights
                p_edges_pruned = weightMatrix.pruneTreshold(p_edges, prune_limit)
                if numpy.sum(p_edges_pruned) == 0 : return partitions
                # normalizing weights
                p_max = numpy.max(p_edges_pruned)
                p_min = numpy.min(p_edges_pruned.nonzero())
                p_zero = p_edges_pruned == 0
                p_edges_norm = (p_edges_pruned - p_min) / (p_max - p_min)
                p_edges_norm[p_zero] = 0
                
                # cluster
                p_partition = cluster(p_edges_norm, indices[partition_mask], split_limit, prune_limit, level + 1, verbose)
            
                # Update partitioning
                partitions[partition_mask] = p_partition + numpy.max(partitions) + 1
    return partitions
```

```python
cluster_weights = weightMatrix.pruneRows(weights_s, 0.05)
#cluster_weights = weightMatrix.pruneTreshold(weights_s, 3)
print("over 0:\t\t\t%.2f" % numpy.sum(cluster_weights > 0))
print("sum over 0:\t\t%.2f" % (numpy.sum(cluster_weights)))
print("edges per sum of edges:\t%.5f" % (numpy.sum(cluster_weights > 0) / float(cluster_weights.size)))
print("edges per vertex:\t%.5f" % (numpy.sum(cluster_weights > 0) / float(cluster_weights.shape[0]*2)))
print(cluster_weights)
cw = cluster_weights
print(numpy.sum(cw - cw.T))
```

```python
partitions = cluster(cluster_weights, indices_s, split_limit = 10000, prune_limit = 1.5, verbose = True)
print("%i partitions" % len(set(partitions)))
```

```python
matches = list(getPartitionMatches(partitions, cluster_weights, indices_s, min_coherence = 0.3))
distances = [matching.matchDistance(ks_s[i].pt, ks_s[j].pt, homography) for i,j in matches]

nb_correct = numpy.sum([1 for d in distances if d < 5])
correct = nb_correct / float(len(matches))
print("Correct: %.2f (%i of %i)" % (correct, nb_correct, len(matches)))
#print(list(matches))
```
