```python
def bothMatch(paths, hom, options = {}) :

    # Returns the position of a match
    def matchPos(i,j) :
        return (features.getPosition(ks[i]), features.getPosition(ks[j]))
    
    # Get options
    match_limit			= options.get("match_limit", 500)
    unique_treshold		= options.get("unique_treshold", 1.0)
    keypoint_type		  = options.get("keypoint_type", "SIFT")
    descriptor_type		= options.get("descriptor_type", "SIFT")
    prune 				 = options.get("prune", False)

    # Get all feature points
    indices, ks, ds = features.getFeatures(paths, keypoint_type = keypoint_type, descriptor_type = descriptor_type)

    # Use cv2's matcher to get matching feature points
    jj,ss,uu = features.bfMatch(descriptor_type, ds, ds, match_same=True)
    
    # See if matches go both ways
    bothways = [(i,j) for i,j in enumerate(jj) if indices[i] == 0 and indices[j] == 1 and jj[j] == i]
    #print(bothways)

    #print(len(bfMatches[0]))
    match_score = [((i,j), matchPos(i,j),ss[i],uu[i]) for i,j in bothways if uu[i] < unique_treshold]
    ind, matches, scores, uniques = zip(*match_score)
    #print(len(uniques))
    #print(uniques)
    
    # Take only the n best
    top_n = numpy.argsort(scores)[0:match_limit]
    matches_top = numpy.array(matches)[top_n]
    
    distances = [matching.matchDistance(ks[i].pt, ks[j].pt, hom) for i,j in numpy.array(ind)[top_n]]
    print(matches_top.shape)
    return matches_top, distances
    

matches, distances = bothMatch(paths, homography, {"unique_treshold" : 0.7, "descriptor_type" : "SIFT", "keypoint_type" : "SIFT"})
```

```python
def mirrorMatch(paths, threshold, options = {}) :

    # Returns the position of a match
    def matchPos(i,j) :
        return (features.getPosition(ks[i]), features.getPosition(ks[j]))

    # Get options
    keypoint_type		= options.get("keypoint_type", "SIFT")
    descriptor_type		= options.get("descriptor_type", "SIFT")

    # Get all feature points
    indices, ks, ds = features.getFeatures(paths, keypoint_type = keypoint_type, descriptor_type = descriptor_type)

    # Use cv2's matcher to get matching feature points
    jj,ss,uu = features.bfMatch(descriptor_type, ds, ds, match_same=True)
    
    # See if matches go both ways
    bothways = [(i,j) for i,j in enumerate(jj) if indices[i] == 0 and indices[j] == 1 and jj[j] == i]
    oneways_1 = [(i,j) for i,j in enumerate(jj) if indices[i] == 0 and indices[j] == 0]
    oneways_2 = [(i,j) for i,j in enumerate(jj) if indices[i] == 1 and indices[j] == 1]

    # Now for each treshold test the uniqueness of the matches
    matches = [matchPos(i,j) for i,j in bothways if uu[i] < threshold]
    matches_im1 = [matchPos(i,j) for i,j in oneways_1 if uu[i] < threshold]
    matches_im2 = [matchPos(i,j) for i,j in oneways_2 if uu[i] < threshold]

    return matches, matches_im1, matches_im2
```

```python
matches, matches_im1, matches_im2 = mirrorMatch(paths, 0.6)
distances = [matching.matchDistance(p1,p2, homography) for p1,p2 in matches]
distances_im1 = [matching.matchDistance(p1,p2, homography) for p1,p2 in matches_im1]
distances_im2 = [matching.matchDistance(p1,p2, homography) for p1,p2 in matches_im2]
display.matchPoints(images[0], images[1], matches, distances, filename="unique-occlusion.png", matches_im1 = matches_im1, dist_im1 = distances_im1, matches_im2 = matches_im2, dist_im2 = distances_im2)
```
