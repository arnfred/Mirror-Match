```python
def getHomographyPath(paths) :
    d = "/".join(paths[0].split('/')[0:-1])
    p1 = paths[0].split('/')[-1].split(".")[0]
    p2 = paths[1].split('/')[-1].split(".")[0]
    return "%s/%s_%s_hom.npy" % (d, p1, p2)
homography_path = getHomographyPath(paths)
with open(homography_path) :
    homography = numpy.loadtxt(homography_path)
```

```python
dt = 15
keypoint_type = "SIFT"
descriptor_type = "SIFT"
```

```python
# Get images
images = map(features.loadImage, paths)

# Get keypoints
indices, ks, ds = features.getFeatures(paths, keypoint_type, descriptor_type)
pos = features.getPositions(ks)
positions = numpy.array(pos)
ds.shape
```

```python
dt = 25
def match_stats(match_pos) :
    dist = [matching.matchDistance(m1, m2, homography) for (m1, m2) in match_pos]
    display.matchPoints(images[0], images[1], match_pos, dist=dist, max_dist = 100)
    nb_correct = len(filter(lambda d : d < dt, dist))
    nb_total = len(match_pos)
    print("Matches: %i, correct: %i (%.2f%%)" % (nb_total, nb_correct, 100.0 * nb_correct / nb_total ) )
```

```python
import ratiomatch
import mirrormatch

threshold_ratio = 0.7
options = {"use_ball_tree":True}
match_pos_r, scores_r, unique_r = mirrormatch.match(paths, options)(threshold_ratio)
match_stats(match_pos_r)
```

```python
import ballmatch

options = {
           "leaf_size": 10,
           "radius_size": 300,
           "dist_threshold": 300,
           "ratio_boost" : 1.2,
           "shuffle_keypoints" : False,
           "group_limit" : 4,
           "match_type" : "radius",
           }
match_fun_b = ballmatch.match(paths, options)
```

```python
threshold_ratio = 0.75
match_pos_b, scores_b, unique_b, group_b = match_fun_b(threshold_ratio)
match_stats(match_pos_b)
```
