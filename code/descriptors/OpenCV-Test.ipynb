# OpenCV SURF #

The idea here is to use the implementation of SURF in opencv to find keypoints and match their descriptors. If a descriptor $d_i$ in an image match a descriptor in another image $d_j$ and $d_j$ also matches to $d_i$ then these two points are selected as matching and returned. To measure if to descriptors match, I measure either the angle between their feature vectors or the hamming distance between them. Which method is used depends on method used to derive the descriptor.

This is different from the method proposed for the sift algorithm which I was using before. Here is a description of that method:

The method used before takes two images and only compares keypoints in one direction. That is, for both images I find all keypoints and their corresponding descriptors. Then I go through the $n$ keypoints of one image ($k_{1_1} ... k_{1_n}$) and for each keypoint $k_{1_i}$ I find the angles between it and all $k_{1_j}$ for ($j \in [1 ... m]$) where $m$ is the amount of keypoints in the second image. Denote by $r_i$ the ratio of the smallest and second smallest angles between $k_{1_i}$ and the keypoints of the second image. If $r_i > c$ for a constant $c$ here defined as $0.6$, the match is deemed unique, and the keypoint $k_{1_i}$ is matched with the keypoint of the second image corresponding to the smallest angle. This method including the value of $c$ is proposed in the sift demo program.

Refs:

* [SURF](https://lirias.kuleuven.be/bitstream/123456789/71383/1/Bay_Tuytelaars_VanGool-surf-eccv06.pdf) (\cite{bay2006surf})
* [SIFT algorithm](http://www.cs.ubc.ca/~lowe/keypoints/) (\cite{lowe2004sift})

```python
import cv2
import os
import fnmatch
from random import sample
import Image

# image folder:
testset = "../../images/testsets/G1M65J/"

# Get a list of all images
images = [testset + f for (_,_,filenames) in os.walk(testset) for f in fnmatch.filter(filenames, "*.png")]

# Pick out all the photos of |Halle Berry (Just to have somebody to test with)
faces = [image for image in images ]# if "Aniston" in image]

# Now pick out two photos at random
[m1, m2] = sample(faces,2)

# Print an image to see what we've got 
# Opencv displays image in BGR, which can be converted to RBG either
# by reverting the last bytes with [:, :, ::-1], or by using the cvtColor function
imshow(cv2.cvtColor(cv2.imread(m1), cv2.COLOR_BGR2RGB))
```

```python
import display
import features as f

# Load images
images = map(f.loadImage, [m1, m2])

# Get keypoints
keypoints = map(lambda i : f.getKeypoints("ORB", i), images)
print(str(len(keypoints[0])) + " keypoints found in im1")

display.keypoints(images[0], keypoints[0])
```

```python
# Get descriptors
descriptors = map(lambda i,k : f.getDescriptors("ORB", i, k), images, keypoints)

# Match
matchPos, scores, uniqueness = f.match(f.hammingDist, descriptors[0], descriptors[1])

# Show result
display.matches(images[0], images[1], keypoints[0], keypoints[1], matchPos)
```

```python
import evaluate

# Filter scores and uniqueness for None items
scores_f = [i for i in scores if i != None]
uniqueness_f = [i for i in uniqueness if i != None]

# Get ranking and reorder both lists in the same order
ranking = numpy.argsort(scores_f)
scores_sorted = [ scores_f[i] for i in ranking]
uniques_sorted = [ uniqueness_f[i] for i in ranking]

# Define n
n = int(len(scores_f)/2.0)

# Get tresholds
score_treshold = sorted(scores_f)[n]
unique_treshold = sorted(uniqueness_f)[n]

```

```python
# Make mask
mask = evaluate.makeMask(lambda (u,s) : u < unique_treshold and s < score_treshold, zip(uniques_sorted, scores_sorted))

# Plot
evaluate.barWithMask(range(len(scores_f)), scores_sorted, mask)
```

In the above plot we see the hamming distances of the matched keypoints ordered in ascending order.

```python
evaluate.barWithMask(range(len(scores_f)), uniques_sorted, mask, color='green')
```

In the above plot we see the ratio of the best and second best match for each matchpoint. They are sorted in the same order as the scores in the plot higher up

```python
# Take the n highest scoring points and plot them
mask = [(u < unique_treshold) for u in uniqueness]
evaluate.matchesWithMask(images, keypoints, matchPos, mask)
```

```python
# Get average scores for selected matchings
averages = [(s,u) for (u,s) in zip(uniques_sorted, scores_sorted) 
                if u <= unique_treshold and s <= score_treshold]

best_scores, best_uniques = zip(*averages)
print("Average score: " + str(mean(best_scores)))
print("Average uniqueness: " + str(mean(best_uniques)))
print("(Lower is better)")
```
