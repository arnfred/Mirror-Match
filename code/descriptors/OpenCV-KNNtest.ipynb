```python
import cv2
import os
import fnmatch
from random import sample
import Image
import features
import display

# image folder:
testset = "../../images/testsets/G1M65J/"

# Get a list of all images
images = [testset + f for (_,_,filenames) in os.walk(testset) for f in fnmatch.filter(filenames, "*.png")]

# Pick out all the photos of |Halle Berry (Just to have somebody to test with)
faces = [image for image in images ]# if "Aniston" in image]

# Now pick out two photos at random
[m1, m2] = sample(faces,2)

# Load images
images = map(features.loadImage, [m1, m2])
```

```python
# Decide on keypoint detector
keypoint_type = features.supported_keypoint_types[4]
print("Keypoint type: " + keypoint_type)

# Get keypoints
keypoints = map(lambda i : features.getKeypoints(keypoint_type, i), images)
print(str(len(keypoints[0])) + " keypoints found in im1")

display.keypoints(images[0], keypoints[0])
```

```python
# Decide on keypoint descriptor
descriptor_type = features.supported_descriptor_types[3]
print("Descriptor type: " + descriptor_type)

# Get descriptors
data = map(lambda i,k : features.getDescriptors(descriptor_type, i, k), images, keypoints)
keypoints, descriptors = zip(*data)
```

```python
print(len(descriptors[0]))
print(len(keypoints[0]))
matchPos, distance, ratio = features.bfMatch(descriptor_type, descriptors[0], descriptors[1])
print(len(matchPos))
treshold = sorted(distance)[17]
matches = [m if distance[i] <= treshold else None
              for (m,i) in zip(matchPos, range(len(matchPos))) ]

print(treshold)

display.matches(images[0], images[1], keypoints[0], keypoints[1],matches)
```
