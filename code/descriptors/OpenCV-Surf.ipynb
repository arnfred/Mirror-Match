# OpenCV SURF #

The idea here is to use the implementation of SURF in opencv to find keypoints and match their descriptors. The method implemented here takes two images and only compares keypoints in one directino. That is, for both images I find all keypoints and their corresponding descriptors. Then I go through the $n$ keypoints of one image ($k_{1_1} ... k_{1_n}$) and for each keypoint $k_{1_i}$ I find the angles between it and all $k_{1_j}$ for ($j \in [1 ... m]$) where $m$ is the amount of keypoints in the second image. Denote by $r_i$ the ratio of the smallest and second smallest angles between $k_{1_i}$ and the keypoints of the second image. If $r_i > c$ for a constant $c$ here defined as $0.6$, the match is deemed unique, and the keypoint $k_{1_i}$ is matched with the keypoint of the second image corresponding to the smallest angle. This method including the value of $c$ is proposed in the sift demo program.

Refs:

* [SURF](https://lirias.kuleuven.be/bitstream/123456789/71383/1/Bay_Tuytelaars_VanGool-surf-eccv06.pdf) (\cite{bay2006surf})
* [SIFT algorithm](http://www.cs.ubc.ca/~lowe/keypoints/) (\cite{lowe2004sift})

```python
import cv2
import os
import fnmatch
from random import sample
import Image

# image folder:
testset = "../../images/testsets/G1M65J/"

# Get a list of all images
images = [testset + f for (_,_,filenames) in os.walk(testset) for f in fnmatch.filter(filenames, "*.png")]

# Pick out all the photos of |Halle Berry (Just to have somebody to test with)
faces = [image for image in images if "Berry" in image]

# Now pick out two photos at random
[m1, m2] = sample(faces,2)
[k1, k2] = [os.path.splitext(image)[0] + ".key" for image in [m1, m2]]
[p1, p2] = [os.path.splitext(image)[0] + ".png" for image in [m1, m2]]
[im1, im2] = [cv2.imread(image) for image in [m1, m2]]

# Print an image to see what we've got 
# Opencv displays image in BGR, which can be converted to RBG either
# by reverting the last bytes with [:, :, ::-1], or by using the cvtColor function

#imshow(im1[:, :, ::-1])
imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
```

Find keypoints and their corresponding descriptors in the two random images

```python
# Get SURF feature detector and extractor
surf = cv2.SURF(200) # cv2.SURF([hessianThreshold[, nOctaves[, nOctaveLayers[, extended[, upright]]]]])
extractor = cv2.DescriptorExtractor_create("SURF")
print(type(surf))

# Try to find keypoints in both images
keypoints_im1 = surf.detect(cv2.cvtColor(im1, cv2.COLOR_BGR2LUV)[:,:,0])
keypoints_im2 = surf.detect(cv2.cvtColor(im2, cv2.COLOR_BGR2LUV)[:,:,0])

print(len(keypoints_im1))
print(len(keypoints_im2))

(keypoints_im1, descriptors_im1) = extractor.compute(im1, keypoints_im1)
(keypoints_im2, descriptors_im2) = extractor.compute(im2, keypoints_im2)
```

Now compare the descriptors of the two images using K nearest neighbors

```python
import numpy

# I need to convert descriptions to float32
type(descriptors_im1)
desc1 = numpy.array(descriptors_im1, dtype = numpy.float32)
desc2 = numpy.array(descriptors_im2, dtype = numpy.float32)

# Train the knn
knn_im1 = cv2.KNearest()
knn_im2 = cv2.KNearest()
knn_im1.train(desc1, numpy.arange(len(descriptors_im1)))
knn_im2.train(descriptors_im2, numpy.arange(len(descriptors_im2)))

# Create test item to match against by looking at the first item of descriptor2
test = numpy.array(desc2[0], dtype = numpy.float32).reshape((1, len(desc1[0])))

# Now find matching pairs of descriptors
r, results, neigh_respo, dists = knn_im1.find_nearest(test,1)
res, dist =  int(results[0][0]), dists[0][0]
print test.shape, results.shape
print res, dist
```

```python
import features as f

def compare2images(path1, path2) :
    # Load images
    images = map(f.loadImage, [path1, path2])
    
    # Get keypoints
    keypoints = map(lambda i : f.getKeypoints("SURF", i), images)
    
    # Get descriptors
    descriptors = map(lambda i,k : f.getDescriptors("SURF", i, k), images, keypoints)
    
    # Match
    matchscores = f.match(descriptors[0], descriptors[1])
    
    # Show result
    sift.plot_matches(images[0], images[1], getPositions(keypoints[0]), getPositions(keypoints[1]), matchscores)

compare2images(p1, p2)
```
