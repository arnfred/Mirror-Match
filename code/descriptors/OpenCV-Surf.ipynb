Load two random images og Halle Berry with opencv

```python
import cv2
import os
import fnmatch
from random import sample
import Image

# image folder:
testset = "../../images/testsets/G1M65J/"

# Get a list of all images
images = [testset + f for (_,_,filenames) in os.walk(testset) for f in fnmatch.filter(filenames, "*.png")]

# Pick out all the photos of |Halle Berry (Just to have somebody to test with)
berries = [image for image in images if "Berry" in image]

# Now pick out two photos at random
[m1, m2] = sample(berries,2)
[k1, k2] = [os.path.splitext(image)[0] + ".key" for image in [m1, m2]]
[p1, p2] = [os.path.splitext(image)[0] + ".png" for image in [m1, m2]]
[im1, im2] = [cv2.imread(image) for image in [m1, m2]]

# Print an image to see what we've got 
# Opencv displays image in BGR, which can be converted to RBG either
# by reverting the last bytes with [:, :, ::-1], or by using the cvtColor function

#imshow(im1[:, :, ::-1])
imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
```

Find keypoints and their corresponding descriptors in the two random images

```python
# Get SURF feature detector and extractor
surf = cv2.SURF() # cv2.SURF([hessianThreshold[, nOctaves[, nOctaveLayers[, extended[, upright]]]]])
extractor = cv2.DescriptorExtractor_create("SURF")


# Try to find keypoints in both images
keypoints_im1 = surf.detect(cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY))
keypoints_im2 = surf.detect(cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY))

(keypoints_im1, descriptors_im1) = extractor.compute(im1, keypoints_im1)
(keypoints_im2, descriptors_im2) = extractor.compute(im2, keypoints_im2)
```

Now compare the descriptors of the two images using K nearest neighbors

```python
import numpy

# I need to convert descriptions to float32
type(descriptors_im1)
desc1 = numpy.array(descriptors_im1, dtype = numpy.float32)
desc2 = numpy.array(descriptors_im2, dtype = numpy.float32)

# Train the knn
knn_im1 = cv2.KNearest()
knn_im2 = cv2.KNearest()
knn_im1.train(desc1, numpy.arange(len(descriptors_im1)))
knn_im2.train(descriptors_im2, numpy.arange(len(descriptors_im2)))

# Create test item to match against by looking at the first item of descriptor2
test = numpy.array(desc2[0], dtype = numpy.float32).reshape((1, len(desc1[0])))

# Now find matching pairs of descriptors
r, results, neigh_respo, dists = knn_im1.find_nearest(test,1)
res, dist =  int(results[0][0]), dists[0][0]
print test.shape, desc1[0].shape
print res, dist
```
