```python
# From here and down is the algorithm in full

# Get images
images = map(features.loadImage, paths)
keypoint_type = "SIFT"
descriptor_type = "SIFT"

# Get keypoints of image 0
indices, ks, ds = features.getFeatures(paths, keypoint_type, descriptor_type)
(pos_im1, pos_im2) = (features.getPositions(ks[indices == 0]), features.getPositions(ks[indices == 1]))
```

```python
# STEP 1: Build the symmetric non-negative n x n matrix M

# get Matches
match_data = features.bfMatch(descriptor_type, ds[indices == 0], ds[indices == 1])

# Get threshold for n best matches
n = 500
t = numpy.sort(match_data[1])[n]

# Get matches and scores
matches = numpy.array([(pos_im1[i], pos_im2[j]) for (i,(j,s,r)) in enumerate(zip(*match_data)) if s < t])
display.matchPoints(images[0], images[1], matches)
scores = numpy.array([s for (j,s,r) in zip(*match_data) if r < t])

# Set sigma
sigma = 50

# Affinity function
def affinity(m1,m2, sigma) :
    d0 = numpy.linalg.norm(m1[0] - m2[0])
    d1 = numpy.linalg.norm(m1[1] - m2[1])
    M_ab = 4.5 - ((d0 - d1) ** 2) / (2*(sigma**2))
    if numpy.abs(d0 - d1) < 3*sigma :
        return M_ab
    else :
        return 0.0

# Construct matrix
M = numpy.zeros((len(matches), len(matches)))
for i,m1 in enumerate(matches) :
    for j,m2 in enumerate(matches) :
        if i == j :
            M[i,j] = 100*(1 - (scores[i] / numpy.max(scores)))
        else :
            M[i,j] = affinity(m1,m2, sigma)
```

```python
# STEP 2:
m = 100

# Get eigenvectors
eigvals, eigvecs = numpy.linalg.eigh(M)
i = eigvals.argsort()[-1]
x_star = eigvecs[:,i]

# For some reason I get an all-negative eigenvector at times.
if x_star[0] < 0 : x_star = x_star * -1
    
# Pick best n
best_m = numpy.array(matches)[numpy.argsort(x_star)[(-1*m):]]
display.matchPoints(images[0], images[1], best_m)
print(numpy.max(x_star))
print(numpy.min(x_star))
print("Best 100 matches picked using geometric constraints")
```
