```python
def get_turntable_path(image_set, object_type, angle, camera_position, pattern_index = 0, turntable_dir = "../../turntable") :
    """ Returns the image path to the image of the set 'image_set' """
    directory = "%s/imageset_%i/%s/%s" % (turntable_dir, image_set, object_type, camera_position)
    file_names = os.listdir(directory)
    file_glob = "img_1-%03i_*_0.JPG" % angle
    image_paths = sorted(fnmatch.filter(file_names, file_glob))
    return "%s/%s" % (directory, image_paths[pattern_index])
```

```python
# Now load two images that we want to calibrate
# Calibration directory
def load_calibration_image(image_set, angle, camera_position, pattern_position, scale = None, turntable_dir = "../../turntable", verbose = False) :
    """ Loads a checkerboard pattern image taken from the same angle and rotation """
    pattern_index = { "flat" : 0, "steep" : 1, "angled" : 2 }[pattern_position]
    image_index = (angle / 5) * 3 + 1 + pattern_index
    image_path = "%s/imageset_%i/%s/%s/calib%i.jpg" % (turntable_dir, image_set, "Calibration", camera_position, image_index)
    if verbose : print(image_path)
    img = cv2.imread(image_path)
    if not scale == None :
        img_scaled = cv2.resize(img, (int(img.shape[1] / scale), int(img.shape[0] / scale)))
        return img_scaled
    else :
        return img
    #return features.loadImage(image_path)
```

```python
# Get calibration points
def get_calibration_points(img, pattern_position) :
    """ Returns the calibration points from an image with the checkerboard pattern """
    if pattern_position == "steep" :
        grid_size = (9, 13)
    else :
        grid_size = (13, 9)
    success, cv_points = cv2.findChessboardCorners(img, grid_size, flags = cv2.CALIB_CB_FILTER_QUADS)
    if not success :
        raise Exception("Can't get lock on chess pattern for img")
    points = numpy.array([[p[0][0], p[0][1]] for p in cv_points])
    #print(points)
    return points# Get calibration points
```

```python
def get_foundamental_matrix(image_set, angles, camera_position, scale = 2.0, return_points = False, verbose = False) :
    """ Returns fundamental matrix between images positions """
    
    # Fetch all images for the angle pair
    img1_flat = load_calibration_image(image_set, angles[0], camera_position[0], "flat", verbose=verbose)
    img2_flat = load_calibration_image(image_set, angles[1], camera_position[1], "flat", verbose=verbose)
    img1_angled = load_calibration_image(image_set, angles[0], camera_position[0], "angled", verbose=verbose)
    img2_angled = load_calibration_image(image_set, angles[1], camera_position[1], "angled", verbose=verbose)
    img1_steep = load_calibration_image(image_set, angles[0], camera_position[0], "steep", verbose=verbose)
    img2_steep = load_calibration_image(image_set, angles[1], camera_position[1], "steep", verbose=verbose)
    
    # Fetch points for all images
    points1_flat = get_calibration_points(img1_flat, "flat")
    points2_flat = get_calibration_points(img2_flat, "flat")
    points1_angled = get_calibration_points(img1_angled, "angled")
    points2_angled = get_calibration_points(img2_angled, "angled")
    points1_steep = get_calibration_points(img1_steep, "steep")
    points2_steep = get_calibration_points(img2_steep, "steep")
    
    # Concatenate point sets
    points1 = numpy.concatenate((points1_flat, points1_angled, points1_steep)) / scale
    points2 = numpy.concatenate((points2_flat, points2_angled, points2_steep)) / scale
    
    # Find fundamental matrix based on points
    F, inliers = cv2.findFundamentalMat(points1, points2, method = cv2.FM_RANSAC)
    
    # return matrix with or without points
    if return_points :
        return F, (points1_flat / scale, points2_flat / scale)
    else :
        return F
```

```python
def epilines(img, points, lines, size = (12, 12), width = 1, name="object") :
    """ Draws a set of epilines and points on an image """
    # Generate figure
    fig = figure(figsize=size)
    imshow(img)
    
    # Get x values
    min_x = 0
    max_x = img.shape[1]
    min_y = 0
    max_y = img.shape[0]
    
    # Limit plot
    xlim(0,max_x)
    ylim(max_y-1,0)
    
    # plot lines
    for l, c in zip(lines, colors.get()) :
        # get line functions
        line_fun = lambda x : (-1 * l[0] * x - l[2]) / (float(l[1]))
        # plot line
        plot([0, max_x], [line_fun(0), line_fun(max_x)], color=c, marker='_', linewidth=width)
    
    # plot points
    for p, c in zip(points, colors.get()) :
        # Plot feature match point
        plot(p[0], p[1], color=c, marker='o', linewidth = width)
        
    ax = pylab.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['bottom'].set_color('none')
    ax.xaxis.set_ticks_position('none')
    
    tick_params(\
        axis='both',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom='off',      # ticks along the bottom edge are off
        top='off',
        left='off',
        right='off',
        labelleft='off',
        labelbottom='off') # labels along the bottom edge are off
            
    savefig("%s.pdf" % name, dpi = 300, bbox_inches='tight', pad_inches=0)
```

```python
def calc_match_distances(matches, img_B, F_AB, F_AC, F_BC, check_threshold) :
    """ Check ground truth for a set of matches given fundamental matrices,
        As proposed in: "Evaluation of Features Detectors and Descriptors 
        based on 3D objects by Pierre Moreels and Peitro Perona.
    """
    
    # Calculate distance between line and 2D-point
    def dist(line, point) :
        return numpy.abs(line.dot([point[0], point[1], 1]))
    
    # return epipolar line
    def get_lines(points, F) :
        return [l[0] for l in cv2.computeCorrespondEpilines(points.reshape(-1, 1, 2), 1, F)]
    
    # Find points that are on l_AB
    points_A = numpy.array([m[0] for m in matches], dtype=numpy.float32)
    points_C = numpy.array([m[1] for m in matches], dtype=numpy.float32)
    lines_AB = get_lines(points_A, F_AB)
    lines_AC = get_lines(points_A, F_AC)
    
    # Collect features from B, so we can check the match there
    feature_pos_B = features.getPositions(features.getKeypoints(img_B))

    for p_A, p_C, l_AB, l_AC in zip(points_A, points_C, lines_AB, lines_AC) :
        
        # Is p_C on l_AC?
        min_dist = dist(l_AC, p_C)
        if min_dist < check_threshold :
            
            # Collect all features in B that are on l_AB:
            points_B = numpy.array([p_B for p_B in feature_pos_B if dist(l_AB, p_B) < check_threshold], dtype=numpy.float32)
            
            # Visualize result
            #epilines(img_B, points_B, [l_AB])
            
            # Check if points match up
            if len(points_B) > 0 :
                lines_BC = get_lines(points_B, F_BC)
                min_dist = max((numpy.min([dist(l_BC, p_C) for l_BC in lines_BC]), min_dist))
                
                # visualize result. note, img_C is global. Make sure it's up to date
                #epilines(img_C, [p_C], lines_BC)
            
        yield min_dist
```

```python
def match_distances(matches, angles, match_object, image_set, check_threshold) :
    
    # Get features in B
    path_B = get_turntable_path(image_set, match_object, angles[0], "Top")
    img_B = features.loadImage(path_B)
    
    # Find fundamental matrices
    F_AC = get_foundamental_matrix(image_set, (angles[0], angles[1]), ("Bottom", "Bottom"), scale = 2.0)
    F_AB = get_foundamental_matrix(image_set, (angles[0], angles[0]), ("Bottom", "Top"), scale = 2.0)
    F_BC = get_foundamental_matrix(image_set, (angles[0], angles[1]), ("Top", "Bottom"), scale = 2.0)
    
    # return distances
    return list(calc_match_distances(matches, img_B, F_AB, F_AC, F_BC, check_threshold))
```

```python
def evaluate(match_fun, angles, image_set, object_type, thresholds, match_count = None, params = {}) :

    # Get distance_threshold
    distance_threshold = params.get("distance_threshold", 5)
    verbose = params.get("verbose", False)
    
    # Get paths to the three images
    path_A = get_turntable_path(image_set, object_type, angles[0], "Bottom")
    path_B = get_turntable_path(image_set, object_type, angles[0], "Top")
    path_C = get_turntable_path(image_set, object_type, angles[1], "Bottom")
    
    # Get match results
    matches = match_fun([path_A, path_C])(1.0)
    
    if match_count != None :
        matches = zip(*[m for m in zip(*matches) if match_count[m[4][0]] > 0])
    
    if verbose :
        print("Found matches")
    
    # Get ground trouth
    distances = match_distances(matches[0], angles, object_type, image_set, distance_threshold)
    
    if verbose :
        print("Found distances")
    
    # Get number of matches given a threshold
    def get_total(t, rs) :
        return int(sum([1 for d, r in zip(distances, rs) if r <= t]))
    
    # Get number of correct matches given a threshold
    def get_correct(t, rs) : 
        return int(sum([1 for d, r in zip(distances, rs) if r <= t and d < distance_threshold]))
    
    # Collect precision per ratio_threshold
    correct = [[get_correct(t, matches[2])] for t in thresholds]
    total = [[get_total(t, matches[2])] for t in thresholds]
    return correct, total
```

```python
# Load keypoints
def keypoints(image_set, object_type, angle, viewpoint) :
    img = features.loadImage(get_turntable_path(image_set, object_type, angle, viewpoint))
    points = features.getPositions(features.getKeypoints(img))
    return numpy.array(points, dtype=numpy.float32)

# return epipolar line
def get_lines(points, F) :
    return [l[0] for l in cv2.computeCorrespondEpilines(points.reshape(-1, 1, 2), 1, F)]

# Calculate distance between line and 2D-point
def dist(line, point) :
    return numpy.abs(line.dot([point[0], point[1], 1]))

def get_correspondences(angles, image_set, object_type, feature_index = None, params = {}) :
    
    # Get distance_threshold
    distance_threshold = params.get("distance_threshold", 5)
    verbose = params.get("verbose", False)

    # Get paths to the three images
    keypoints_A = keypoints(image_set, object_type, angles[0], "Bottom")
    keypoints_B = keypoints(image_set, object_type, angles[0], "Top")
    keypoints_C = keypoints(image_set, object_type, angles[1], "Bottom")
    
    # Find fundamental matrices
    F_AC = get_foundamental_matrix(image_set, (angles[0], angles[1]), ("Bottom", "Bottom"), scale = 2.0)
    F_AB = get_foundamental_matrix(image_set, (angles[0], angles[0]), ("Bottom", "Top"), scale = 2.0)
    F_BC = get_foundamental_matrix(image_set, (angles[0], angles[1]), ("Top", "Bottom"), scale = 2.0)
    
    # For every point in A find the corresponding lines in B and C
    lines_AB = get_lines(keypoints_A, F_AB)
    lines_AC = get_lines(keypoints_A, F_AC)
    
    # Collect matches
    matches = [[]] * keypoints_A.shape[0] 
    
    # For every epiline in B and C corresponding to a point in A
    for i, (p_A, l_AB, l_AC) in enumerate(zip(keypoints_A, lines_AB, lines_AC)) :
        
        # Find all points on the line in B and C
        points_B = numpy.array([p_B for p_B in keypoints_B if dist(l_AB, p_B) < distance_threshold], dtype=numpy.float32)
        points_C = numpy.array([p_C for p_C in keypoints_C if dist(l_AC, p_C) < distance_threshold], dtype=numpy.float32)
        
        if i == feature_index :
            epi_size = (7,7)
            epilines(img_A, [p_A], [], size = epi_size, width = 2, name = "Frog_A")
            epilines(img_B, points_B, [l_AB], size = epi_size, width = 2, name = "Frog_B")
            epilines(img_C, points_C, [l_AC], size = epi_size, width = 2, name = "Frog_C")
            if len(points_B) > 0 :
                epilines(img_C, points_C, get_lines(points_B, F_BC), size = epi_size, width = 2, name = "Frog_C_lines")
        
        # For every point in B on l_AB, see if there is a point in C on l_AC that lies on the epipolar line of p_B in image C: l_BC
        if len(points_B) > 0 and len(points_C) > 0 :
            
            # Get distances from every point in C on line l_AC to every line in C corresponding to a point in B on line l_AB
            get_distances = lambda line : [(dist(line, p_C), p_C) for p_C in points_C]
            distances = features.flatten([get_distances(l_BC) for l_BC in get_lines(points_B, F_BC)])
            
            # Count how many potential matches there are
            matches[i] = [(p_A, p_C) for d, p_C in distances if d < distance_threshold]
            
    return matches
```
