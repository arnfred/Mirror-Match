# Selecting Subjects #

In the folder 'Images' there is a large set of images containing faces of famous people.
They have been downloaded from the [faces in the wild](http://vis-www.cs.umass.edu/lfw/) 
database which consists of images containing the face of a person and a corresponding filename
labeled to match the name of the given person.

## Steps needed ##

To use these faces for this project I will need to select faces for which we have more than $n$ photos, where $n$ is set to $10$ to begin with. To obtain partial faces I'll rotate and crop the images according to the procedure descriped in [Partial Face Recognition: Alignment-Free Approach](http://ieeexplore.ieee.org/xpls/abs_all.jsp?arnumber=6117573) (\cite[p. 8]{liao2011}).

```python
from os import listdir

# The file holding a list of all the personal image folders
# and how many images they contain
imagecount_file = "../../images/person-appearances"

# The number of pictures per person we need as a minimum
min_pictures = 10

# Read in the person appearances file
with open(imagecount_file) as f:
    content = f.readlines()
    
# echo out each folder that contains more than ten pictures
people = [line.split()[0] for line in content if int(line.split()[1]) >= min_pictures]

# The final list of people suitable for experimentation
print(str(len(people)) + " people with " + str(min_pictures) + " or more pictures:\n")
print("\n".join(people))
```
