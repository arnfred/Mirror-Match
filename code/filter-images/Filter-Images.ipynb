# Selecting Subjects #

In the folder 'Images' there is a large set of images containing faces of famous people.
They have been downloaded from the [faces in the wild](http://vis-www.cs.umass.edu/lfw/) 
database which consists of images containing the face of a person and a corresponding filename
labeled to match the name of the given person.

To use these faces for this project I will need to select faces for which we have more than $n$ photos, where $n$ is set to $10$ to begin with. 

```python
# The file holding a list of all the personal image folders
# and how many images they contain
image_folder = "../../images/"
imagecount_file = image_folder + "person-appearances"


# The number of pictures per person we need as a minimum
min_pictures = 10 #n
nb_people = 60 #m


# Read in the person appearances file
with open(imagecount_file) as f:
    content = f.readlines()
    
    
# Create a list of all the folders containing at least min_pictures images of a person
def getPeople() : 
    return [line.split()[0] for line in content if int(line.split()[1]) >= min_pictures]


# Print of the final list of people suitable for experimentation
people = getPeople()
print(str(len(people)) + " people with " + str(min_pictures) + " or more pictures:\n")
print("\n".join(people))
```

```python
from random import choice, sample

# Get random person with more than $n$ images
def getRandPeople(m) :
    return sample(people,m)
```

```python
from os import listdir
from os.path import isfile, join
from scipy import misc
# Get random image of random person
def getRandImages(name,n) :
    # Get image path
    path = image_folder + "all/" + name + "/"
    
    # Get a list of images associated with this person
    images = [str(f) for f in listdir(path) if isfile(join(path,f))]
    
    # Select a single image
    image_paths = [path + image for image in sample(images,n)]
    
    return image_paths

# Print a random image to illustrate that the method works fine
name = getRandPeople(1)[0]
imshow(imread(getRandImages(name,1)[0]))
print(name.replace('_',' '))
```

## Creating testset album ##

Each testset is stored in an album with a name that is generated randomly and consists of 6 characters

```python
from string import ascii_uppercase, digits
from os.path import isdir
from os import makedirs

# Position of the testsets folder
testset_folder = image_folder + "testsets/"


# Generate a folder with a random name
def foldername_generator(size=6, chars=ascii_uppercase + digits):
    condition = True
    while condition:
        name = ''.join(choice(chars) for x in range(size))
        condition = isdir(testset_folder + name)
    return testset_folder + name


# Now create a folder with this name
foldername = foldername_generator()
makedirs(foldername)


# Print out the new name
print(foldername)
```

## Picking a group of random images ##

Now that images can be loaded, I'll pick out between $1$ and $n$ images from $m$ persons and save them to the folder created above

```python
from random import randint

# how many images we use
def nb_images() : return randint(1, min_pictures)


# Copy images to testset folder
def get_images() :
    path = image_folder + "all/"
    return [image for name in getRandPeople(nb_people) 
                                        for image in getRandImages(name, nb_images())]

# Get images
images = get_images()
print(images)
```

```python
from shutil import copy

# Move images to folder
def copy_images(images, folder) :
    for image in images :
        name = image.split("/")[-1]
        print("copying " + name + " to " + folder)
        copy(image, folder)

# Do it for our particular case
copy_images(images, foldername)
```
