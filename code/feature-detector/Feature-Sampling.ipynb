```python
def getRandKeypoint(image, margin=15, size=32) :
    x = random.random()*(image.shape[1]-2*margin)+margin
    y = random.random()*(image.shape[0]-2*margin)+margin
    return cv2.KeyPoint(x,y,_size=size)
print(getRandKeypoint(images[0]).pt)
```

```python
def getRandKeypoints(image, n = 1000, margin=16, size=32) :
    return [getRandKeypoint(image, margin=margin, size=size) for _ in range(0,n)]
```

```python
def getSurprisePoints(image, n = 500, step=100, iterations = 10) :
    # Initialize our sampling set
    ks_init = getRandKeypoints(image, step)
    ks_out, ds_collected = features.getDescriptors("BRIEF", image, ks_init)
    ks_collected = numpy.array(ks_out)
    
    # Initialize binary descriptor matrix
    bin_collected = numpy.array(numpy.unpackbits(ds_collected), dtype=numpy.bool)
    bin_collected.shape = (ds_collected.shape[0], ds_collected.shape[1] * 8)

    for i in range(iterations) :
        print
        # Initialize keypoints and descriptors
        ks_init = getRandKeypoints(image, step)
        ks_current, ds_current = features.getDescriptors("BRIEF", image, ks_init)
        
        # Calculate the probability of a descriptor, P(X) = #{X_i = 1 for all X_i \in X} / |X|
        desc_prop = numpy.sum(bin_collected, axis=0) / float(bin_collected.shape[0])
        desc_prop_inv = 1 - desc_prop
        desc_prop_log = numpy.log(desc_prop)
        desc_prop_inv_log = numpy.log(desc_prop_inv)
        
        # Take the log probability of every single descriptor with naive bayesian assumption
        prop_collected = desc_prop_log*bin_collected + desc_prop_inv_log*numpy.invert(bin_collected)
        prop_log_collected = numpy.sum(prop_collected, axis=1)
        prop_log_collected_mean = numpy.mean(prop_log_collected)
        print("Prop log mean: %.4f" % prop_log_collected_mean)
        
        # Calculate probability for descriptors of this iteration
        bin_current = numpy.array(numpy.unpackbits(ds_current), dtype=numpy.bool)
        bin_current.shape = (ds_current.shape[0], ds_current.shape[1] * 8)
        prop_current = numpy.log(desc_prop)*bin_current + numpy.log(desc_prop_inv)*numpy.invert(bin_current)
        prop_log_current = numpy.sum(prop_current, axis=1)
        
        # Take out all descriptors below the mean
        mask_current = prop_log_current < prop_log_collected_mean
        print("New keypoints added: %i" % numpy.sum(mask_current))
        #ks_collected = numpy.concatenate((ks_collected, numpy.array(ks_current)[mask_current]))
        #ds_collected = numpy.concatenate((ds_collected, ds_current[mask_current]))
        #bin_collected = numpy.concatenate((bin_collected, bin_current[mask_current]))
        bin_collected = numpy.concatenate((bin_collected, bin_current))
        ks_collected = numpy.concatenate((ks_collected, numpy.array(ks_current)))
        ds_collected = numpy.concatenate((ds_collected, ds_current))
    
    return_indices = numpy.argsort(prop_log_collected)
    print(prop_log_collected)
    print(return_indices)
    ds_return = ds_collected[return_indices[-n::]]
    ks_return = ks_collected[return_indices[-n::]]
    return ks_return, ds_return


ks, ds = getSurprisePoints(images[0], n = 1, step=50, iterations=2)
ks_pos = features.getPositions(ks)
print(ds.shape)
display.compareKeypoints(images[0], images[1],ks_pos , [])
```

```python
from itertools import product
import pylab

# works only for quadratic image right now
def getPropHeatMap(image, size=42) :
    ks_init = [cv2.KeyPoint(x,y,_size=size) for (x,y) in product(range(image.shape[0]), repeat=2)]
    #ks_init = [cv2.KeyPoint(x,50,_size=size) for x in range(0,128)]
    ks, ds  = features.getDescriptors("BRIEF", image, ks_init)
    print(ds.shape)
    # Initialize binary descriptor matrix
    
    # get descriptors in binary
    bin_collected = numpy.array(numpy.unpackbits(ds), dtype=numpy.bool)
    bin_collected.shape = (ds.shape[0], ds.shape[1] * 8)
    
    # Calculate the probability of a descriptor, P(X) = #{X_i = 1 for all X_i \in X} / |X|
    desc_prop = (numpy.sum(bin_collected, axis=0) + 1) / (float(bin_collected.shape[0]) + 1)
    desc_prop_inv = 1.00001 - desc_prop
    desc_prop_log = numpy.log(desc_prop)
    desc_prop_inv_log = numpy.log(desc_prop_inv)
    
    # Take the log probability of every single descriptor with naive bayesian assumption
    prop_collected = desc_prop_log*bin_collected + desc_prop_inv_log*numpy.invert(bin_collected)
    prop_log_collected = numpy.sum(prop_collected, axis=1)
    
    # Normalize
    max_prop = numpy.max(prop_log_collected)
    min_prop = numpy.min(prop_log_collected)
    prop_norm = (prop_log_collected - min_prop) / (max_prop - min_prop)
    
    # Create heatmap data
    prop_map = numpy.ones(image.shape)
    for i,k in enumerate(ks) :
        [x,y] = k.pt
        #print("x: %i, y: %i" % (x,y))
        prop_map[image.shape[0]-y,x] = prop_norm[i]
    
    #print(prop_map[:,50])
    
    # Draw heatmap
    pylab.pcolor(prop_map)
    pylab.colorbar()
    pylab.show()       
    
    
    
print(numpy.product(images[0].shape))
getPropHeatMap(images[0])
```

```python
# works only for quadratic image right now
def getDistanceHeatMap(image, size=42) :
    ks_init = [cv2.KeyPoint(x,y,_size=size) for (x,y) in product(range(image.shape[0]), repeat=2)]
    #ks_init = [cv2.KeyPoint(x,50,_size=size) for x in range(64,128)]
    ks, ds  = features.getDescriptors("BRIEF", image, ks_init)
    print(ds.shape)
    # Initialize binary descriptor matrix
    
    # get descriptors in binary
    bin_collected = numpy.array(numpy.unpackbits(ds), dtype=numpy.bool)
    bin_collected.shape = (ds.shape[0], ds.shape[1] * 8)
    
    # Calculate the 'mean' descriptor
    desc_prop = numpy.array(2*numpy.sum(bin_collected, axis=0) / bin_collected.shape[0], dtype=numpy.bool)
    
    # Calculate the distance to the mean descriptor for all descriptors
    prop_collected = numpy.bitwise_xor(desc_prop, bin_collected)
    prop_sum_collected = numpy.sum(prop_collected, axis=1)
    
    # Normalize
    max_prop = numpy.max(prop_sum_collected)
    min_prop = numpy.min(prop_sum_collected)
    prop_norm = (prop_sum_collected - min_prop) / float(max_prop - min_prop)
    
    # Create heatmap data
    prop_map = numpy.ones(image.shape)
    for i,k in enumerate(ks) :
        [x,y] = k.pt
        #print("x: %i, y: %i" % (x,y))
        prop_map[128-y,x] = prop_norm[i]
    
    max_index = numpy.argmax(prop_norm)
    k = ks[max_index]
    #print(prop_map[:,50])
    
    # Draw heatmap
    pylab.pcolor(prop_map)
    pylab.colorbar()
    #pylab.xlim(image.shape[1])
    display.removeDecoration()   
    
    
    
print(numpy.product(images[0].shape))
getDistanceHeatMap(images[0])
```

```python
import weightMatrix

# works only for quadratic image right now
def getClosestNHeatMap(image, size=42, n = 5) :
    
    # Function for getting the distance of the closest n neighbors
    def getMeanOfClosestN(row) :
        closest_n = argsort(row)[0:n]
        return numpy.mean(row[closest_n])
    
    ks_init = [cv2.KeyPoint(x,y,_size=size) for (x,y) in product(range(image.shape[0]), repeat=2)]
    #ks_init = [cv2.KeyPoint(x,50,_size=size) for x in range(64,128)]
    ks, ds  = features.getDescriptors("BRIEF", image, ks_init)
    print(ds.shape)
    
    # Initialize binary descriptor matrix
    dist = weightMatrix.hammingDist(ds)
    
    # For each descriptor get the neighbors
    dist_closest_n = [getMeanOfClosestN(row) for row in dist]
        
    # Normalize
    max_dist = numpy.max(dist_closest_n)
    min_dist = numpy.min(dist_closest_n)
    dist_closest_n_norm = (dist_closest_n - min_dist) / float(max_dist - min_dist)
    
    # Create heatmap data
    dist_map = numpy.ones(image.shape)
    for i,k in enumerate(ks) :
        [x,y] = k.pt
        #print("x: %i, y: %i" % (x,y))
        dist_map[image.shape[0]-y,x] = dist_closest_n_norm[i]
    
    # Draw heatmap
    pylab.pcolor(dist_map)
    pylab.colorbar()
    #pylab.xlim(image.shape[1])
    display.removeDecoration()   
```
