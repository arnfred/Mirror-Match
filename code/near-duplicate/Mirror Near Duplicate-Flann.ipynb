```python
scores_norm_count = duplicates.norm_counts(scores, modifier = 1000)
duplicates.heatmap(scores_norm_count, labels, cmap=cm.jet)
```

```python
scores_capped = duplicates.cap(scores_norm_count)
duplicates.heatmap(scores_capped, labels, cmap=cm.jet)
```

```python
scores_trimmed = duplicates.trim(scores_capped, margins = (0.3, 0.7), verbose = True)
duplicates.heatmap(scores_trimmed, labels, cmap=cm.jet)
```

```python
s = scores_trimmed.copy()
for i in range(s.shape[0]) :
    s[i,i] = 1
scores_squared = (s).dot(s)
duplicates.heatmap(scores_squared, labels, cmap=cm.jet)
```

```python
scores_squared_norm = duplicates.norm_scores(scores_squared)
duplicates.heatmap(scores_squared_trimmed, labels, cmap=cm.jet)
```

```python
calif_ground_truth = numpy.load("../near-duplicate/resources/california3_non_binary10_groundtruth.npy")
duplicates.heatmap(calif_ground_truth, labels)
```

```python
def max_dist(scores, dist) :
    scores_short = numpy.zeros(scores.shape)
    for i in range(scores_squared_norm.shape[0]) :    
        for j in range(scores_squared_norm.shape[0]) :
            if numpy.abs(i - j) < dist :
                scores_short[i, j] = scores[i, j]
    return scores_short

# Set diagonal
for i in range(scores_squared_norm.shape[0]) :
    scores_squared_trimmed[i,i] = 1
    calif_ground_truth[i, i] = 1

# Remove photos far apart in time
scores_squared_trimmed = duplicates.trim(scores_squared_norm, margins = (0.0, 0.7), verbose = True)
s = max_dist(scores_squared_trimmed, 55)

duplicates.get_F1(s, calif_ground_truth)
```
