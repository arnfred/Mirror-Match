```python
def matches_fun(threshold) :
    return zip(*[(p,s,u,i) for p,s,u,i in zip(*matches) if u <= threshold])
scores, counts, labels = duplicates.get_scores(matches_fun, files, threshold = 0.8)
```

```python
scores_norm_count = duplicates.norm_counts(scores, counts, modifier = 100)
#duplicates.print_scores(scores_norm_count, labels)
scores_capped = duplicates.cap(scores_norm_count, nb_sd = 0.5, modifier = 100)
#duplicates.print_scores(scores_capped, labels) #,use_float=True)
duplicates.heatmap(scores_capped, labels, cmap=cm.jet)
```

```python
scores_norm = duplicates.norm_scores(scores_capped, amount = 10, min_val = 1, verbose = True)
duplicates.heatmap(scores_norm, labels, cmap=cm.jet)
```

```python
scores_trimmed = duplicates.trim(scores_norm, margins = (0.2,0.8), verbose = True)
duplicates.heatmap(scores_trimmed, labels, cmap=cm.jet)
```

```python
s = duplicates.set_diag(scores_trimmed, 1)
scores_squared = s.dot(s)
duplicates.heatmap(scores_squared, labels, cmap=cm.jet)
#start = 300
#end = 330
#duplicates.print_scores(scores_squared[start:end][:,start:end], labels[start:end])
```

```python
scores_squared_norm = duplicates.trim(scores_squared, margins = (0.3, 0.55), verbose = True)
duplicates.heatmap(scores_squared_norm, labels, cmap = cm.jet)
```
