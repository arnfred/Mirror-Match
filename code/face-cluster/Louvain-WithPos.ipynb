```python
indices, keypoints, descriptors = features.getFeatures(paths, keypoint_type="ORB", size=32)
print("Number of keypoints: %i" % (len(keypoints)))
print(set([k.size for k in keypoints]))
print(set([k.octave for k in keypoints]))


images = map(features.loadImage, paths)
labels = map(features.getLabel, paths)
print(labels)

```

```python
pos_0 = map(features.getPosition, keypoints[numpy.array(indices)==0])
pos_1 = map(features.getPosition, keypoints[numpy.array(indices)==1])
display.compareKeypoints(images[0], images[1], pos_0, pos_1)
```

```python
# Get weights
full_weights = weightMatrix.init(descriptors)
score_weights = weightMatrix.pruneTreshold(full_weights, 40)
cluster_weights = weightMatrix.pruneHighest(score_weights, 10)
```

```python
# Get positions
x_list,y_list = zip(*map(features.getPosition, keypoints))
x = numpy.array(x_list)
y = numpy.array(y_list)

x_outer = numpy.outer(x,x)
y_outer = numpy.outer(y,y)
x_sq = numpy.outer(x*x,numpy.ones(x.shape))
y_sq = numpy.outer(y*y,numpy.ones(y.shape))
#dist_mat = numpy.sqrt(x_sq + x_sq.T + y_sq + y_sq.T - 2*(x_outer + y_outer))
dist_mat = (x_sq + x_sq.T + y_sq + y_sq.T - 2*(x_outer + y_outer))
min_d = numpy.min(dist_mat)
max_d = numpy.max(dist_mat)
dist = (dist_mat-min_d) / (max_d - min_d)
dist_reversed = (1 - dist)*0.7
dist_reversed[dist==0] = 0.0
```

```python
# Construct matrix where inter-image edges are subtracted their distance
dist_split = numpy.zeros(dist.shape)

ind = numpy.array(indices)
im0_mask = ind == 0
im1_mask = ind == 1
for i,row in enumerate(dist_split) :
    row[im0_mask] = dist[i][im0_mask]*im0_mask[i]
    row[im1_mask] = dist[i][im1_mask]*im1_mask[i]
```

```python
cluster_geom_full = numpy.zeros(full_weights.shape)
for i,row in enumerate(cluster_geom_full) :
    row[im0_mask] = dist_reversed[i][im0_mask]*im0_mask[i] + full_weights[i][im0_mask]*im1_mask[i]
    row[im1_mask] = dist_reversed[i][im1_mask]*im1_mask[i] + full_weights[i][im1_mask]*im0_mask[i]
#print(full_weights)
#print
#print(cluster_geom_full)
cluster_geom = weightMatrix.pruneHighest(cluster_geom_full, 5)
#print
#print(cluster_geom[3])
#print(cluster_geom[])
```

```python
geom_scale = 5.0 # Some number deciding how much spaced out keypoints are punished
cluster_dist = cluster_weights - (geom_scale * dist_split)
cluster_dist[cluster_geom<0] = 0
```

```python
# Normalize score_weights
zeros = score_weights == 0
non_z = score_weights != 0
max_c = numpy.max(score_weights)
min_c = numpy.min(score_weights) - 0.001
score_w_norm = (score_weights - min_c) / (max_c - min_c)
score_w_norm[zeros] = 0.0

# Get graphs
weights = cluster_geom
cluster_graph = featureGraph.init(weights)
score_graph = featureGraph.init(score_w_norm)
```

```python
import louvain

#partition = louvain.cluster(cluster_geom, True)
partition = louvain.cluster(weights, True)
p = featureGraph.setPartition(cluster_graph, partition)
```

```python
featureGraph.setFeatureIndices(cluster_graph, indices)
featureGraph.setFeatureIndices(score_graph, indices)
display.draw_graph(cluster_graph, p)
```

```python
# Show on image
c = [28]
featureGraph.setPositions(cluster_graph, keypoints)

one_cluster = featureGraph.getPartition(cluster_graph, p, c)
display.graph_on_images(one_cluster, images, p)
```

```python
# Score the clusters conventionally
scores_geom, partition_indices_geom = scoring.scoreWeights(score_weights, partition, numpy.array(indices), verbose=True)
print("score: %0.4f" % (sum(scores_geom)))
```

```python
# Score the clusters conventionally
scores, partition_indices = scoring.scoreWeights_2(score_weights, partition, numpy.array(indices), verbose=True)
print("score: %0.4f" % (sum(scores)))
```

```python
if len(scores_geom) > 0 :
    tg = featureGraph.make_trait_graph(cluster_graph, p, partition_indices_geom, scores_geom)
    display.trait_graph(tg, images)
```

```python
# Get the geometric multiplier
geom_multiplier = scoring.geometryMultiplier(partition, partition_indices, numpy.array(indices), keypoints)

# Get the certainty factor
certainty_factor = scoring.certaintyFactor(len(partition_indices))
print("")
print("Certainty: %.2f" % certainty_factor)

# Print new scores
print("")
for s, m, p_i in zip(scores, geom_multiplier, partition_indices) : print("[%i]\to: %+.4f, g: %+.4f, s: %+.4f" % (p_i,s,m,s*m))
score_sum = numpy.sum([s*m for s,m in zip(geom_multiplier, scores)])

# Get final score by multiplying certainty
final_score = score_sum * certainty_factor
print("")
print("Final score: %0.4f" % (sum(final_score)))
```
